```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report,confusion_matrix
```

```python
data = pd.read_csv("IMDB Dataset.csv",encoding="latin")
data.head()
```

```python
data.sentiment.value_counts()
```

```python
import seaborn as sns
data.groupby('sentiment').describe
sns.countplot(data=data, x='sentiment')
```

```python
import matplotlib.pyplot as plt
fig3, ax = plt.subplots(1, figsize=(10, 7))
labels=['Negative','Positive']
sizes=[data.sentiment.value_counts()]
colors = ['Orange', 'red']
explode = (0, 0)  # explode 1st slice
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
ax.set_title("Value count of label i.e. Negative and Positive calculated")
plt.legend()
plt.axis('equal')
plt.show()
```

```python


x_train,x_test,y_train,y_test = train_test_split(data.review, data.sentiment, test_size=0.5, random_state=37)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Using Naive Bayes Model

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer()
vectorized_text = vectorizer.fit_transform(x_train)
```

```python
from sklearn.naive_bayes import MultinomialNB

classifier = MultinomialNB()
classifier.fit(vectorized_text,y_train)
```

```python
x_test_vector=vectorizer.transform(x_test)
final_prediction_NB=classifier.predict(x_test_vector)
```

```python
count = 0
for i in range(len(y_test)):
    if y_test.iloc[i] != final_prediction_NB[i]:
        count += 1
print('Total number of test cases', ((len(y_test))))
print('Number of wrong of predictions', count)
```

```python
print(classification_report( y_test,final_prediction_NB))
```

```python
print(sns.heatmap(confusion_matrix(y_test,final_prediction_NB),annot=True,fmt=''))
```

# Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
classifier_logistic =  LogisticRegression(random_state=0)
classifier_logistic.fit(vectorized_text,y_train)
```

```python
final_prediction_LR=classifier_logistic.predict(x_test_vector)
```

# ANN with embedding layers and precised model

```python
X=data.review
Y=data.sentiment
le=LabelEncoder()
Y=le.fit_transform(Y)
```

```python
print(X.shape)
print(Y.shape)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.5, random_state = 41)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
max_length = 120
vocab_size = 10000
embeding_dim = 16 
trunc_type = 'post'
oov_tok="<OOV>"
```

```python
from keras.preprocessing.text import Tokenizer 
from keras.preprocessing import sequence
tokenizer=Tokenizer(num_words=vocab_size,oov_token=oov_tok)
tokenizer.fit_on_texts(x_train)
word_index = tokenizer.word_index
sequences = tokenizer.texts_to_sequences(x_train)
padded_sequence=sequence.pad_sequences(sequences,padding='post',truncating=trunc_type,maxlen=max_length)
```

```python
testing_sequence = tokenizer.texts_to_sequences(x_test)
padded_testing = sequence.pad_sequences(testing_sequence,padding='post',truncating=trunc_type,maxlen=max_length)
```

```python
print(padded_sequence.shape)
print(y_train.shape)
print(padded_testing.shape)
print(y_test.shape)
```

```python
import tensorflow as tf
```

```python
model=tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size,embeding_dim,input_length=max_length),
    tf.keras.layers.GRU(3),
    tf.keras.layers.Dense(6,activation='relu'),
#     tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1,activation='sigmoid')
])

model.compile(loss='binary_crossentropy',metrics=['acc'],optimizer='adam')
model.summary()
```

```python
history = model.fit(padded_sequence,y_train,epochs=30,validation_data=(padded_testing,y_test),verbose=2)
```

```python
import matplotlib.pyplot as plt
def plot_graphs(history, string):
    plt.plot(history.history[string])
    plt.plot(history.history['val_'+string])
    plt.xlabel("Epochs")
    plt.ylabel(string)
    plt.legend([string, 'val_'+string])
    plt.show()
plot_graphs(history, "acc")
plot_graphs(history, "loss")
```

# Bidirectional convolutional layer

```python
from tensorflow.keras import regularizers
model1 = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size,embeding_dim,input_length = max_length),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64)),
    tf.keras.layers.Dense(64,activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1,activation='sigmoid')
])

model1.compile(loss='binary_crossentropy',optimizer=tf.keras.optimizers.Adam(lr=0.001),metrics=['acc'])
model1.summary()
```

```python
earlystop = tf.keras.callbacks.EarlyStopping(monitor = 'val_loss', 
                          patience = 3,
                          verbose = 1,
                          restore_best_weights = True)
history1 = model1.fit(padded_sequence,y_train,epochs=10,validation_data=(padded_testing,y_test),verbose=2,callbacks=[earlystop])
```

```python
plot_graphs(history1, "acc")
plot_graphs(history1, "loss")
```
